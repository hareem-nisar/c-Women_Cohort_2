```python
from matplotlib import image
from matplotlib import pyplot as pp
import numpy as np

img1 = image.imread('rainbow.jpg')
pp.imshow(img1,  cmap="gray")# wont effect the image
pp.show()


# separate R, G, B images


redCh = img1.copy()
#redCh[# specify [#rows, #cols, #channels]] = Pixel Value
redCh[:,:,1:3] = 0      #setting G, B to 0
pp.imshow(redCh)
pp.show()

greenCh = img1.copy()
greenCh[:,:,[0,2]] = 0      #setting R = 0
#greenCh[:,:,2] = 0      #setting B = 0
pp.imshow(greenCh)
pp.show()

blueCh = img1.copy()
#redCh[# specify [#rows, #cols, #channels]] = Pixel Value
blueCh[:,:, 0:2] = 0      #setting R, G to 0
pp.imshow(blueCh)
pp.show()

print(img1.dtype) #uint8
# mixing R, B
#newColor = img1.copy()
#newColor[:,:, 0:2] = 0  
newColor = redCh*0.5 + blueCh*0.5
print(newColor.dtype)
newColor = newColor.astype("uint8")
pp.imshow(newColor)
pp.show()
print(newColor.dtype)


# get an avergage gray image from an RGB image
g = (img1[:,:,0] + img1[:,:,1] + img1[:,:,2])/3   # Gray_pixel_intensity = R+G+B/3
g = g.astype("uint8")
pp.imshow(g, cmap="gray")
pp.show()


g2 = (img1[:,:,0] + img1[:,:,1] )/57   # Gray_pixel_intensity = R+G+B/3
g2 = g2.astype("uint8")
pp.imshow(g2, cmap="RdPu")
pp.show()
```

```python
# blending two images
img1 = image.imread('rainbow.jpg')
img2 = image.imread("love.jpg")

print(img1.shape)
print(img2.shape)

img1 = img1[0:200, 0:200, :]
img2 = img2[0:200, 0:200, :]

print(img1.shape)
print(img2.shape)


blended_img = (img1 * 0.3 + img2 * 0.7).astype(np.uint8)   # Blending them in
pp.imshow(blended_img)
```
